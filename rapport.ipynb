*Question 3*

La signature d’un morceau repose sur des triplets de la forme :

$$
(t_i - t_a,\ f_a,\ f_i)
$$

Cette représentation est **invariante par translation temporelle**.  
En effet, si le signal est décalé dans le temps d’une quantité \(T\), les instants deviennent \(t_a + T\) et \(t_i + T\), mais leur différence reste inchangée :

$$
(t_i + T) - (t_a + T) = t_i - t_a
$$
Ainsi, les hashcodes ne dépendent pas du temps absolu mais uniquement des relations locales entre points du spectrogramme. Cela permet de reconnaître un extrait indépendamment de sa position dans le morceau.


*Question 5*

Les correspondances entre les hashcodes d’un extrait et d’un morceau sont représentées sous forme de points 
$$
(t_a - \tilde{t}_a)
$$


Cas où l’extrait correspond au morceau :
  Les points sont alignés selon une droite de pente 1.  
  Cela traduit un décalage temporel constant entre l’extrait et le morceau.

Cas où l’extrait ne correspond pas :
  Les points sont répartis de manière aléatoire, sans structure particulière.


*Question 6*

On considère les différences :

$$
(t_a,\ \tilde{t}_a)
$$

- Si l’extrait correspond au morceau :
  L’histogramme présente un pic marqué, correspondant au décalage temporel entre l’extrait et le morceau.

- Cas où l’extrait ne correspond pas :
  L’histogramme est diffus, sans pic dominant.


*Question 7*

On utilise l’histogramme des offsets pour déterminer si un extrait correspond à un morceau.

On considère qu’il y a correspondance si le plus grand pic est significativement supérieur aux autres :


$$
\max(\mathrm{hist}) > \mathrm{critere} \times \mathrm{second\_max}
$$
